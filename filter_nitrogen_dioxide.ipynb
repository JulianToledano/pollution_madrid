```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/ene_mo18.csv', delimiter=';')
df.head()
```

```python
nitrogen_dioxide = df['MAGNITUD'] == 8
df = df[nitrogen_dioxide]
df.tail()
```

```python
# Super ugly :S
# Calculate de mean using just V values. lamda will be great.
for index, row in df.iterrows():
    mean_values = []
    for i in range(1, 25):
        if row['V'+str(i).zfill(2) ] == 'V':
            df.loc[index, 'MEDIA'] = float("NaN")
```

```python
cols = [c for c in df.columns if c[0] == 'V']
df = df.drop(labels=cols, axis=1)
df['MEDIA'] = df.loc[: , "H01":].mean(axis=1)
df['MEDIA_MADRUGADA'] = df.loc[: , "H01":"H08"].mean(axis=1)
df['MEDIA_MAÑANA'] = df.loc[: , "H08":"H16"].mean(axis=1)
df['MEDIA_TARDE'] = df.loc[: , "H16":"H24"].mean(axis=1)
df.head()
```
